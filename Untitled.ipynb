```python
# Dependencies
import tweepy
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
from config import consumer_key, consumer_secret, access_token, access_token_secret

#print "Current date and time using strftime:"
time_now = datetime.datetime.now().strftime("%d %B %Y / %H:%M EST")

# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# Establish a variable for the number of tweets to retrieve from each data source...useful during testing

number_tweets_per_source = 100
```

```python
# Create list of the news sources
news_source_list = ['@BBCWorld',  '@CBSNews', '@CNN', '@FoxNews', '@nytimes'] # @BBCNews is just UK news, so chose @BBCWorld
```

```python
# Create list of column names for the dataframe
cols = ['Source', 'Text', 'Date', 'Compound',  'Positive', 'Neutral', 'Negative']
```

```python
# Pull last 100 tweets from BBC, CBS, CNN, Fox,  NY Times

# Create an empty dataframe

tweet_df = pd.DataFrame()

for source in news_source_list: # loop through all 5 news sources, updating the dataframe for each source
    tweet_data_list = [] # create empty list to hold data for each tweet; reset to empty for each new source
    df_temp = [] # create temp df which is filled with the data from a single news source
    tweet_100 = api.search(source, count=number_tweets_per_source)
    print(source)
    # loop through each tweet recovered from the news source... in this case 100 tweets
    for tweet in tweet_100['statuses']: 
        results = analyzer.polarity_scores(tweet['text']) # Vader sentiment analysis on the tweet text
        tweet_data_list.append([source, tweet['text'], tweet['created_at'], results['compound'], results['pos'],
                                results['neu'], results['neg']])
        df_temp = pd.DataFrame(tweet_data_list, columns=cols) 
    tweet_df = tweet_df.append(df_temp, ignore_index=True)

tweet_df.head(15)   
```

```python
tweet_df.count()
```

```python
# Change df index to Source to facilitate data extraction by news source

tweet_df.set_index('Source', inplace=True)
```

```python
# Extract df to csv file
file_name = 'Media Sentiment Analysis.csv'

tweet_df.to_csv(file_name, encoding='utf-8')

tweet_df.head(12)
```

```python
# Create individual dataframes for the sentiment values for each new source (i.e. BBC, CBS, etc) for use in scatter plot

# Create a list that represents the # of tweets for each news source (range 1 to 100 in this case)
number_list = list(range(1,number_tweets_per_source+1))  

# Convert this number list into a series
se = pd.Series(number_list)

# Add this numbers series to each of the new dataframes
bbc_sentiment_df = tweet_df.loc[['@BBCWorld'],['Compound']]
bbc_sentiment_df['Tweets Ago'] = se.values

cbs_sentiment_df = tweet_df.loc[['@CBSNews'],['Compound']]
cbs_sentiment_df['Tweets Ago'] = se.values

cnn_sentiment_df = tweet_df.loc[['@CNN'],['Compound']]
cnn_sentiment_df['Tweets Ago'] = se.values

fox_sentiment_df = tweet_df.loc[['@FoxNews'],['Compound']]
fox_sentiment_df['Tweets Ago'] = se.values

nyt_sentiment_df = tweet_df.loc[['@nytimes'],['Compound']]
nyt_sentiment_df['Tweets Ago'] = se.values
```

# Create first plot - scatter plot of sentiment analysis

```python
# Make scatter plot
mylabels = ['BBC', 'CBS', 'CNN', 'Fox', 'NY Times']

plt.xlim(number_tweets_per_source + 5 ,-5)
plt.ylim(-1.1,1.1)

plt.xlabel('Tweets Ago')
plt.ylabel('Tweet Polarity')
plt.title(f'Sentiment Analysis of Media Tweets \n {time_now}')

plt.scatter(bbc_sentiment_df['Tweets Ago'],bbc_sentiment_df['Compound'],edgecolors='black')
plt.scatter(cbs_sentiment_df['Tweets Ago'],cbs_sentiment_df['Compound'],edgecolors='black')
plt.scatter(cnn_sentiment_df['Tweets Ago'],cnn_sentiment_df['Compound'],edgecolors='black')
plt.scatter(fox_sentiment_df['Tweets Ago'],fox_sentiment_df['Compound'],edgecolors='black')
plt.scatter(nyt_sentiment_df['Tweets Ago'],nyt_sentiment_df['Compound'],edgecolors='black')
plt.grid(color='lightgrey', linestyle='-', linewidth=.5, alpha=.5)
plt.legend(loc=(1.04, .65), labels=mylabels)
plt.savefig('Sentiment Analysis.png')

plt.show()
```

# Create 2nd Plot - Bar Plot Visualizing Overall Sentiments

```python
# Create array with the values of the averages for the Compound sentiments

myvalues_array = np.array([bbc_sentiment_df['Compound'].mean(),
        cbs_sentiment_df['Compound'].mean(),
        cnn_sentiment_df['Compound'].mean(),
        fox_sentiment_df['Compound'].mean(),
        nyt_sentiment_df['Compound'].mean()])

mylabels_array = np.array(['BBC', 'CBS', 'CNN', 'Fox', 'NY Times'])

#colors = np.array(['c','gold','g','r','purple'])

print(myvalues_array)
print(mylabels_array)
```

```python
# Create bar plot

for i in range(5): # loop through all 5 news sources, plotting a bar for each
    if myvalues_array[i] < 0: 
        plt.bar(mylabels_array[i], myvalues_array[i], color='lightcoral', edgecolor = 'black', align='center', width = 1)
        plt.text(i-.25, -.02, str(round(myvalues_array[i],2)), color = 'black')
    else:
        plt.bar(mylabels_array[i], myvalues_array[i], color='palegreen', edgecolor = 'black', align='center', width = 1)
        plt.text(i-.2, .02, str(round(myvalues_array[i],2)), color = 'black')
        
plt.ylabel('Tweet Polarity')
plt.title(f'Overall Media Sentiment based on Twitter \n {time_now}')
plt.savefig('Overall Sentiment Bar Chart.png')
plt.show()
```
